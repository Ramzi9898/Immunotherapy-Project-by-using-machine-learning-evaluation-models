```python
import matplotlib.pyplot as plt
from sklearn.metrics import roc_curve, auc

# True labels and predicted probabilities for KNN
y_true_knn = [0, 1, 1, 0, 1, 0, 1, 0, 0, 1]
pred_proba_knn = [0.2, 0.8, 0.6, 0.3, 0.9, 0.1, 0.7, 0.4, 0.2, 0.6]

# True labels and predicted probabilities for SVM
y_true_svm = [0, 1, 1, 0, 1, 0, 1, 0, 0, 1]
pred_proba_svm = [0.3, 0.7, 0.4, 0.2, 0.6, 0.1, 0.8, 0.5, 0.3, 0.7]

# True labels and predicted probabilities for logistic regression
y_true_logistic = [0, 1, 1, 0, 1, 0, 1, 0, 0, 1]
pred_proba_logistic = [0.1, 0.9, 0.5, 0.4, 0.7, 0.2, 0.6, 0.3, 0.1, 0.8]

# True labels and predicted probabilities for neural network
y_true_neural = [0, 1, 1, 0, 1, 0, 1, 0, 0, 1]
pred_proba_neural = [0.4, 0.6, 0.8, 0.1, 0.9, 0.3, 0.7, 0.2, 0.5, 0.6]

# True labels and predicted probabilities for regularization
y_true_regularization = [0, 1, 1, 0, 1, 0, 1, 0, 0, 1]
pred_proba_regularization = [0.2, 0.8, 0.6, 0.3, 0.9, 0.1, 0.7, 0.4, 0.2, 0.6]

# Compute the false positive rate (FPR) and true positive rate (TPR) for each technique
fpr_knn, tpr_knn, _ = roc_curve(y_true_knn, pred_proba_knn)
fpr_svm, tpr_svm, _ = roc_curve(y_true_svm, pred_proba_svm)
fpr_logistic, tpr_logistic, _ = roc_curve(y_true_logistic, pred_proba_logistic)
fpr_neural, tpr_neural, _ = roc_curve(y_true_neural, pred_proba_neural)
fpr_regularization, tpr_regularization, _ = roc_curve(y_true_regularization, pred_proba_regularization)

# Compute the Area Under the ROC Curve (AUC) for each technique
auc_knn = auc(fpr_knn, tpr_knn)
auc_svm = auc(fpr_svm, tpr_svm)
auc_logistic = auc(fpr_logistic, tpr_logistic)
auc_neural = auc(fpr_neural, tpr_neural)
auc_regularization = auc(fpr_regularization, tpr_regularization)

# Plot ROC curves
plt.plot(fpr_knn, tpr_knn, label='KNN (AUC = {:.2f})'.format(auc_knn))
plt.plot(fpr_svm, tpr_svm, label='SVM (AUC = {:.2f})'.format(auc_svm))
plt.plot(fpr_logistic, tpr_logistic, label='Logistic Regression (AUC = {:.2f})'.format(auc_logistic))
plt.plot(fpr_neural, tpr_neural, label='Neural Network (AUC = {:.2f})'.format(auc_neural))
plt.plot(fpr_regularization, tpr_regularization, label='Regularization (AUC = {:.2f})'.format(auc_regularization))

# Set labels and title
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('Receiver Operating Characteristic (ROC) Curves')

# Set legend
plt.legend(loc='lower right')

# Show the plot
plt.show()
```
